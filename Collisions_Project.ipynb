```python
import requests
import pandas as pd

base_url = "https://ckan0.cf.opendata.inter.prod-toronto.ca"

dataset_id = "police-annual-statistical-report-traffic-collisions"

package_url = f"{base_url}/api/3/action/package_show"
params = {"id": dataset_id}
response = requests.get(package_url, params=params)
package = response.json()

resource = next((r for r in package['result']['resources'] if r['datastore_active']), None)

if resource:
    resource_id = resource["id"]
    all_records = []
    limit = 10000
    offset = 0

    while True:
        data_url = f"{base_url}/api/3/action/datastore_search"
        query = {
            "id": resource_id,
            "limit": limit,
            "offset": offset
        }
        res = requests.get(data_url, params=query).json()
        batch = res["result"]["records"]
        all_records.extend(batch)
        print(f"Fetched {len(batch)} records (Total so far: {len(all_records)})")

        if len(batch) < limit:
            break

        offset += limit

    df = pd.DataFrame(all_records)
    print(f"\n Total records fetched: {len(df)}")
```

```python
import pandas as pd
import numpy as np

df_clean = df.copy()

columns_to_drop = [
    '_id', 'OCC_MONTH', 'OCC_DOW', 'DIVISION',
    'FTR_COLLISIONS', 'HOOD_158', 'NEIGHBOURHOOD_158', 'geometry'
]
df_clean.drop(columns=columns_to_drop, inplace=True, errors='ignore')

df_clean['OCC_DATE'] = pd.to_datetime(df_clean['OCC_DATE'], unit='ms', errors='coerce')  # Convert from UNIX time
df_clean['YEAR'] = df_clean['OCC_DATE'].dt.year
df_clean['MONTH'] = df_clean['OCC_DATE'].dt.month
df_clean['DAY'] = df_clean['OCC_DATE'].dt.day
df_clean['WEEKDAY'] = df_clean['OCC_DATE'].dt.day_name()

df_clean['FATALITIES'] = pd.to_numeric(df_clean['FATALITIES'], errors='coerce').fillna(0).astype(int)

def yes_no_to_bool(val):
    return 1 if str(val).strip().upper() == 'YES' else 0

for col in ['INJURY_COLLISIONS', 'FTR_COLLISIONS', 'AUTOMOBILE', 'MOTORCYCLE', 'PASSENGER', 'BICYCLE', 'PEDESTRIAN']:
    if col in df_clean.columns:
        df_clean[col] = df_clean[col].apply(yes_no_to_bool)

df_clean = df_clean[(df_clean['LAT_WGS84'] != 0) & (df_clean['LONG_WGS84'] != 0)]
print(df_clean.info())
df_clean.head()
```

```python
df_clean['LAT'] = df_clean['LAT_WGS84'].round(5)
df_clean['LON'] = df_clean['LONG_WGS84'].round(5)

grouped = df_clean.groupby(['LAT', 'LON']).agg({
    'FATALITIES': 'sum',
    'INJURY_COLLISIONS': 'sum',
    'OCC_DATE': 'count'
}).reset_index()

grouped.rename(columns={'OCC_DATE': 'TOTAL_COLLISIONS'}, inplace=True)

grouped['DANGER_SCORE'] = (
    grouped['TOTAL_COLLISIONS'] * 1 +
    grouped['INJURY_COLLISIONS'] * 2 +
    grouped['FATALITIES'] * 5
)

grouped['LAT'] = pd.to_numeric(grouped['LAT'], errors='coerce')
grouped['LON'] = pd.to_numeric(grouped['LON'], errors='coerce')

grouped = grouped[
    (grouped['LAT'].abs() > 0.1) &
    (grouped['LON'].abs() > 0.1)
]

top30 = grouped.sort_values(by='DANGER_SCORE', ascending=False).head(30).reset_index(drop=True)

print(top30)
```

```python
import folium
top30['RANK'] = (top30.index + 1).astype(int)

toronto_map = folium.Map(location=[43.651070, -79.347015], zoom_start=12)

for _, row in top30.iterrows():
    folium.CircleMarker(
        location=[row['LAT'], row['LON']],
        radius=0.5 + row['DANGER_SCORE'] * 0.02,
        color='red',
        fill=True,
        fill_opacity=0.7,
        popup=(
            f"Collisions: {row['TOTAL_COLLISIONS']}<br>"
            f"Injuries: {row['INJURY_COLLISIONS']}<br>"
            f"Fatalities: {row['FATALITIES']}"
        )
    ).add_to(toronto_map)
    folium.map.Marker(
        [row['LAT'], row['LON']],
        icon=folium.DivIcon(
            html=f"""<div style="font-size: 12pt; color: black; font-weight: bold">#{row['RANK']}</div>"""
        )
    ).add_to(toronto_map)

toronto_map
```

```python
# Sort intersections by total collisions descending
grouped_sorted = grouped.sort_values(by='TOTAL_COLLISIONS', ascending=False)

# Number of intersections representing the top 20%
top_20pct_count = int(len(grouped_sorted) * 0.20)

# Collisions from top 20% intersections
collisions_top20 = grouped_sorted.head(top_20pct_count)['TOTAL_COLLISIONS'].sum()

# Total collisions
total_collisions = grouped_sorted['TOTAL_COLLISIONS'].sum()

# Percentage contribution
pct_collisions = (collisions_top20 / total_collisions) * 100

print(f"Top 20% of intersections account for {pct_collisions:.2f}% of all collisions.")
```
