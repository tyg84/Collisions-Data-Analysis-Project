```python
import requests
import pandas as pd

base_url = "https://ckan0.cf.opendata.inter.prod-toronto.ca"

dataset_id = "police-annual-statistical-report-traffic-collisions"

package_url = f"{base_url}/api/3/action/package_show"
params = {"id": dataset_id}
response = requests.get(package_url, params=params)
package = response.json()

resource = next((r for r in package['result']['resources'] if r['datastore_active']), None)

if resource:
    resource_id = resource["id"]
    all_records = []
    limit = 10000
    offset = 0

    while True:
        data_url = f"{base_url}/api/3/action/datastore_search"
        query = {
            "id": resource_id,
            "limit": limit,
            "offset": offset
        }
        res = requests.get(data_url, params=query).json()
        batch = res["result"]["records"]
        all_records.extend(batch)
        print(f"Fetched {len(batch)} records (Total so far: {len(all_records)})")

        if len(batch) < limit:
            break

        offset += limit

    df = pd.DataFrame(all_records)
    print(f"\n Total records fetched: {len(df)}")
```

```python
import pandas as pd
import numpy as np

df_clean = df.copy()

columns_to_drop = [
    '_id', 'OCC_MONTH', 'OCC_DOW', 'DIVISION',
    'FTR_COLLISIONS', 'HOOD_158', 'NEIGHBOURHOOD_158', 'geometry'
]
df_clean.drop(columns=columns_to_drop, inplace=True, errors='ignore')

df_clean['OCC_DATE'] = pd.to_datetime(df_clean['OCC_DATE'], unit='ms', errors='coerce')  # Convert from UNIX time
df_clean['YEAR'] = df_clean['OCC_DATE'].dt.year
df_clean['MONTH'] = df_clean['OCC_DATE'].dt.month
df_clean['DAY'] = df_clean['OCC_DATE'].dt.day
df_clean['WEEKDAY'] = df_clean['OCC_DATE'].dt.day_name()

df_clean['FATALITIES'] = pd.to_numeric(df_clean['FATALITIES'], errors='coerce').fillna(0).astype(int)

def yes_no_to_bool(val):
    return 1 if str(val).strip().upper() == 'YES' else 0

for col in ['INJURY_COLLISIONS', 'FTR_COLLISIONS', 'AUTOMOBILE', 'MOTORCYCLE', 'PASSENGER', 'BICYCLE', 'PEDESTRIAN']:
    if col in df_clean.columns:
        df_clean[col] = df_clean[col].apply(yes_no_to_bool)

df_clean = df_clean[(df_clean['LAT_WGS84'] != 0) & (df_clean['LONG_WGS84'] != 0)]
print(df_clean.info())
df_clean.head()
```

```python
df_clean['LAT'] = df_clean['LAT_WGS84'].round(5)
df_clean['LON'] = df_clean['LONG_WGS84'].round(5)

grouped = df_clean.groupby(['LAT', 'LON']).agg({
    'FATALITIES': 'sum',
    'INJURY_COLLISIONS': 'sum',
    'OCC_DATE': 'count',
    'AUTOMOBILE': 'sum',
    'MOTORCYCLE': 'sum',
    'PASSENGER': 'sum',
    'BICYCLE': 'sum',
    'PEDESTRIAN': 'sum'
}).reset_index()

grouped.rename(columns={'OCC_DATE': 'TOTAL_COLLISIONS'}, inplace=True)

grouped['DANGER_SCORE'] = (
    grouped['TOTAL_COLLISIONS'] * 1 +
    grouped['INJURY_COLLISIONS'] * 5 +
    grouped['FATALITIES'] * 20
)

grouped['LAT'] = pd.to_numeric(grouped['LAT'], errors='coerce')
grouped['LON'] = pd.to_numeric(grouped['LON'], errors='coerce')

grouped = grouped[
    (grouped['LAT'].abs() > 0.1) &
    (grouped['LON'].abs() > 0.1)
]

top30 = grouped.sort_values(
    by='DANGER_SCORE',
    ascending=False
).head(30).reset_index(drop=True)

print(top30)
```

```python
import geopandas as gpd

# Load ward boundaries
wards = gpd.read_file("City Wards Data - 4326.geojson").to_crs("EPSG:4326")

# Convert grouped table to GeoDataFrame
gdf_points = gpd.GeoDataFrame(
    grouped,
    geometry=gpd.points_from_xy(grouped['LON'], grouped['LAT']),
    crs="EPSG:4326"
)

# Spatial join
gdf_joined = gpd.sjoin(
    gdf_points,
    wards[['AREA_SHORT_CODE', 'geometry']],
    how='left',
    predicate='within'
)

# Rename and clean up
gdf_joined.rename(columns={'AREA_SHORT_CODE': 'WARD_NUM'}, inplace=True)
gdf_joined.drop(columns=['index_right', 'geometry'], inplace=True, errors='ignore')

# Final dataset (same data + ward number)
grouped_with_ward = gdf_joined
# I will manually handle 39 rows of missing data
na_rows = grouped_with_ward[grouped_with_ward['WARD_NUM'].isna()]
na_rows.to_csv("wards_to_fix_manually.csv", index=False)

```

```python
#output stuff to fix
import pandas as pd

# Ensure we have a stable row identifier
grouped_with_ward = grouped_with_ward.reset_index(drop=True)
grouped_with_ward['ROW_ID'] = grouped_with_ward.index

# Extract only rows with missing ward
to_fix = grouped_with_ward[grouped_with_ward['WARD_NUM'].isna()].copy()

# Export for manual correction
to_fix[['ROW_ID', 'LAT', 'LON', 'WARD_NUM']].to_csv(
    "wards_to_fix_manually.csv",
    index=False
)

print(f"Exported {len(to_fix)} rows to wards_to_fix_manually.csv")
```

```python
#import corrected wards and apply them ---

fixes = pd.read_csv("wards_to_fix_manually_corrected.csv")

# Sanity checks (very important)
assert 'ROW_ID' in fixes.columns, "ROW_ID column missing"
assert fixes['WARD_NUM'].notna().all(), "Some WARD_NUM values are still missing"

# Apply fixes using ROW_ID (NO floating point joins)
grouped_with_ward.loc[
    fixes['ROW_ID'],
    'WARD_NUM'
] = fixes['WARD_NUM'].values

# Track provenance
grouped_with_ward['WARD_SOURCE'] = 'auto'
grouped_with_ward.loc[fixes['ROW_ID'], 'WARD_SOURCE'] = 'manual'

# Final safety check
assert grouped_with_ward['WARD_NUM'].isna().sum() == 0, "Still missing wards!"

# Normalize WARD_NUM: "01" -> "1", keep NaN intact
grouped_with_ward['WARD_NUM'] = (
    grouped_with_ward['WARD_NUM']
    .astype(str)
    .str.strip()
    .replace({'nan': None, '': None})
    .apply(lambda x: str(int(x)) if x is not None else None)
)


# Export final CSV (overwrite existing file used by HTML)
grouped_with_ward.to_csv(
    "collisions_with_wards.csv",
    index=False
)

print("Final CSV exported: collisions_with_wards.csv")
```
